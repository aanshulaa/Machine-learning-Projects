# Implementation of linear regression

```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
#importing test and training data
train_file_path = input("Enter the file path for the training dataset (e.g., train.csv): ")
test_file_path = input("Enter the file path for the testing dataset (e.g., test.csv): ")
```

```python
train_df = pd.read_csv(train_file_path)
```

```python
train_df.head()
```

```python
X_train = train_df[['LotFrontage', 'LotArea', 'MSSubClass', 'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']]
y_train = train_df['SalePrice']
```

```python
# Distribution of the target variable (SalePrice)
plt.figure(figsize=(10, 6))
sns.histplot(train_df['SalePrice'], bins=30, kde=True)
plt.title('Distribution of SalePrice')
plt.show()
```

```python
# Pairplot for a subset of numerical features
numerical_features_subset = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', '1stFlrSF']
sns.pairplot(train_df[numerical_features_subset])
plt.show()
```

```python
numeric_columns = train_df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr()


plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title("Pearson Correlation Heatmap")
plt.show()
```

```python
# Display the number of null values in each column
null_counts = train_df.isnull().sum()
columns_with_nulls = null_counts[null_counts > 0]
print(columns_with_nulls)
```

```python
# Create a pipeline with an imputer and a linear regression model
model = make_pipeline(SimpleImputer(strategy='mean', fill_value=0), LinearRegression())
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Load the test dataset
test_df = pd.read_csv(test_file_path)
```

```python
# Extract features from the test set
X_test = test_df[['LotFrontage', 'LotArea', 'MSSubClass',
                  'OverallQual', 'OverallCond', 'YearBuilt', 'YearRemodAdd', 'BsmtFinSF1', 
                  'BsmtFinSF2', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'BsmtFullBath', 
                  'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
                  'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 
                  'PoolArea', 'MiscVal', 'MoSold', 'YrSold']]
```

```python
predictions = model.predict(X_test)
```

```python
# Add the predictions to the test DataFrame
test_df['predicted_SalePrice'] = predictions
```

```python
# Display the test DataFrame with predictions using linear regression
print(test_df[['Id', 'predicted_SalePrice']])
```

```python
# Evaluate the model on the training set
y_train_pred = model.predict(X_train)
mse_train = mean_squared_error(y_train, y_train_pred)
print(f'Training Mean Squared Error: {mse_train}')
```

```python
# Plot the predicted vs actual prices on the training set
plt.scatter(y_train, y_train_pred)
plt.xlabel('Actual Sale Prices (Training)')
plt.ylabel('Predicted Sale Prices (Training)')
plt.title('Actual Sale Prices vs Predicted Sale Prices (Training)')
plt.show()
```
